```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xlrd
pd.options.display.max_columns = 50
```

```python
from jupyterthemes import jtplot
jtplot.style(theme='monokai', context='notebook', ticks=True, grid=False)
```

```python
# Importing Data Source

df = pd.read_csv(r'C:\Users\fikri\Desktop\pyproj\vidio.com\213.csv',error_bad_lines=False)
print(df.shape)
print(df.columns)
print(df.head())
print(df.info(verbose=True))
```

```python
print(df.info(verbose=True))
print(df.head())
```

```python
#List unique value certain columns

list = (['is_login','platform','playback_location','browser_name','app_name','referrer_group','completed','player_name','os_name','is_premium','city','content_type','film_title','season_name','genre_name','stream_type','category_name'])
for col_name in list:
    print('\nUnique Values Count Variable',col_name)
    print(df[col_name].value_counts())
```

```python
#drop unnecessary

droplist = ['referrer','average_bitrate','bitrate_range','utm_source','utm_medium','utm_campaign','flash_version','os_version'
            ,'browser_version','app_version','city','player_name']
df=df.drop(droplist, axis=1)
```

```python
#null value check!

for col in df:
    print("row kosong pada kolom {}, {}".format(col, df[col].isnull().sum()))
```

```python
# make date_visit and hour visiting 
df['play_time'] = pd.to_datetime(df.play_time).dt.tz_localize(None)
df['date_visit']= df['play_time'].dt.date
df['hour']=df['play_time'].dt.hour
print(df['date_visit'].head())
print(df['hour'].head())
```

```python
#plotting date vs total visiting
df.groupby('date_visit')['date_visit'].count().plot.line()
plt.rcParams['figure.figsize']=(20, 10)
plt.style.use('dark_background')
plt.xticks(df.date_visit.unique(),size=20)
plt.xlabel('Tanggal', size=18, labelpad=10)
plt.ylabel('Jumlah Pengunjung', size=18, labelpad=10)
plt.show()
```

```python
#plotting date vs total play duration
df_datevplay = df.groupby('date_visit')['play_duration'].sum().reset_index().sort_values(by='date_visit', ascending = True)#
plt.rcParams['figure.figsize']=(40, 10)
sns.lineplot(x='date_visit', y='play_duration', data=df_datevplay,palette = 'gnuplot')
plt.xticks(df_datevplay['date_visit'],size=20)
plt.xlabel('Tanggal', size=14, labelpad=10)
plt.ylabel('Total durasi streaming', size=18, labelpad=10)
plt.show()
```

```python
#plotting hour vs total visiting
df.groupby('hour')['hour'].count().plot.line()
plt.xticks(df.hour.unique())
plt.xlabel('Jam Kunjungan', size=18, labelpad=10)
plt.ylabel('Jumlah Pengunjung', size=18, labelpad=10)
plt.show()
```

```python
#premium eda
plt.rcParams['figure.figsize']=(20, 30)
plt.style.use('dark_background')
sns.countplot(y='is_premium' , data=df ,palette = 'gnuplot')
plt.show()
```

```python
#referrer group
plt.rcParams['figure.figsize']=(20, 30)
plt.style.use('dark_background')
ref_grp = df[df['referrer_group'] != 'others']
ref_grp1 = ref_grp['referrer_group'].value_counts().head(10).reset_index()
print(ref_grp1)
sns.barplot(x='index',y='referrer_group' , data=ref_grp1 ,palette = 'gnuplot')
plt.xlabel('Referrer Group', size=18, labelpad=10)
plt.xlabel('Total', size=18, labelpad=10)
plt.xticks(rotation=30, fontsize=12)
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=12, color='white', ha='center', va='bottom')
plt.show()
```

```python

#platform EDA
df_platform = df['platform'].reset_index()
df_platform = df_platform.replace(['tv-android','tv-tizen','tv-webos'],['SmartTV','SmartTV','SmartTV'])
plt.rcParams['figure.figsize']=(20, 30)
plt.style.use('dark_background')
sns.countplot(y='platform' , data=df_platform ,palette = 'gnuplot')
plt.show()
```

```python
#Content EDA
plt.rcParams['figure.figsize']=(10, 20)
plt.style.use('dark_background')
sns.countplot(y='content_type' , data=df ,palette = 'gnuplot', order=df['content_type'].value_counts().index)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.xlabel('count',size=16)
plt.ylabel('content type',size=16)
plt.show()
```

```python
#film vod
plt.rcParams['figure.figsize']=(20, 10)
film_vod = df[df['content_type'] == 'vod']
film_vod1 = film_vod['film_title'].value_counts().head(10).reset_index()
print(film_vod1)
sns.barplot(x='index',y='film_title' , data=film_vod1 ,palette = 'gnuplot')
plt.tick_params(axis='both', which='major', labelsize=16)
plt.ylabel('count',size=16)
plt.xlabel('Judul Film', size=18, labelpad=10)
plt.xticks(rotation=30, fontsize=12)
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=12, color='white', ha='center', va='bottom')
plt.show()
```

```python
#content vod

content_vod = film_vod['title'].value_counts().head(10).reset_index()
print(content_vod)
sns.barplot(x='index',y='title' , data=content_vod ,palette = 'gnuplot')
plt.xticks(rotation=90, fontsize=12)
plt.xlabel('Judul Konten', size=18, labelpad=10)
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=12, color='white', ha='center', va='bottom')
plt.show()
```

```python
#live streaming
ls = df[df['content_type'] == 'livestreaming']
ls = ls['title'].value_counts().head(10).reset_index()
print(ls)
sns.barplot(x='index',y='title' , data=ls ,palette = 'gnuplot')
plt.xticks(rotation=90, fontsize=28)
plt.yticks( fontsize=28)
plt.xlabel('Judul Konten', size=18, labelpad=10)
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=12, color='white', ha='center', va='bottom')
plt.show()
```

```python
## completed ratio
df_comp = df[df['completed'] == True]
complete = df_comp.groupby('title')['completed'].count().reset_index().sort_values(by='completed', ascending = False).head(10)
df_total = df['title'].value_counts().reset_index()
df_total = df_total.rename(columns={'title': 'total'})
df_complete_join = complete.set_index('title').join(df_total.set_index('index')).reset_index()
df_complete_join['%completed'] = (df_complete_join['completed']/df_complete_join['total'])*100
print(df_complete_join)


x_com = df_complete_join['title']
y1_com = df_complete_join['completed']
y2_com = df_complete_join['%completed']
fig, ax1 = plt.subplots()
plt.xticks(rotation=90, fontsize=20)
ax2 = ax1.twinx()
ax1.bar(x_com, y1_com, color='b')
ax2.plot(x_com, y2_com, color='w')
ax2.set_ylim([0, 100])
plt.show()
```
