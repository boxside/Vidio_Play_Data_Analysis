```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import xlrd
pd.options.display.max_columns = 50
```

```python
# Importing Data Source

df = pd.read_csv(r'C:\Users\fikri\Desktop\pyproj\vidio.com\213.csv',error_bad_lines=False)
print(df.shape)
print(df.columns)
print(df.head())
print(df.info(verbose=True))
```

```python
# make date_visit and hour visiting 
df['play_time'] = pd.to_datetime(df.play_time).dt.tz_localize(None)
df['date_visit']= df['play_time'].dt.date
df['hour']=df['play_time'].dt.hour
print(df['date_visit'].head())
print(df['hour'].head())
```

```python
#top 10 visitor by play_duration/day
watcher = df[['date_visit','hash_watcher_id','play_duration']]
day_watcher1 = watcher.groupby(['hash_watcher_id'])['date_visit'].nunique().sort_values(ascending=False).reset_index()
day_watcher1=day_watcher1.rename(columns={'date_visit': 'jumlah_hari'})

sumplay_watcher= watcher.groupby(['hash_watcher_id'])['play_duration'].sum().reset_index().sort_values(by='play_duration', ascending = False)
merged = sumplay_watcher.set_index('hash_watcher_id').join(day_watcher1.set_index('hash_watcher_id')).reset_index()
merged['play/day'] = merged['play_duration']/merged['jumlah_hari']
merged1 = merged[['hash_watcher_id','play/day']].sort_values(by='play/day', ascending = False)
print(merged1.head(10))

merged1.head(10).to_excel('playday1.xlsx',engine='xlsxwriter')
```
